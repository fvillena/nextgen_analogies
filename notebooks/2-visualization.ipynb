```python
import matplotlib.pyplot as plt
import json
```

```python
with open("../reports/PlanTL-GOB-ES--roberta-base-biomedical-clinical-es.json", "r") as f:
    roberta_biomedical_clinical = json.load(f)
```

```python
with open("../reports/PlanTL-GOB-ES--roberta-base-bne.json", "r") as f:
    roberta_base = json.load(f)
```

```python
for rela, accuracy in roberta_biomedical_clinical.items():
    plt.plot(range(1, len(accuracy)+1), accuracy, label=rela)
plt.xlabel("k")
plt.ylabel("accuracy")
# plot legend outside of the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
